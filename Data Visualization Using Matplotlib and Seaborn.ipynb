```python
import pandas as pd
```

```python

df_sales = pd.read_excel("D:/Python/linechart.xlsx")
#df_sales.set_index('Quarter', inplace=True)
df_sales
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 4))
for column in df_sales.columns:
    plt.plot(df_sales.index, df_sales[column], marker='o', label=column)

# Customize the chart
plt.title('Appliance Sales Trend Over Quarters')
plt.xlabel('Quarter')
plt.ylabel('Units Sold')
plt.grid(False)
plt.legend()
plt.tight_layout()
```

```python
total_sales = df_sales[["Fridge","Dishwasher","Washing Machine"]].sum()
total_sales
```

```python
total_sales.index
```

```python
plt.pie(total_sales, labels=total_sales.index, autopct='%1.1f%%',explode=(0.1,0,0),shadow=True, startangle=90)

plt.show()
```
